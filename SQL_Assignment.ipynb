***Question 2 : What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.***

SQL constraints are rules applied to columns or tables that limit the type of data that can be inserted, updated, or deleted, ensuring the accuracy, consistency, and reliability of the data in the database
These are divided into 6 types, mainly;
a)Primary Key
b)Foreign Key
C)Not Null
d)Unique
e)Check
f)Default
a)Primary Key-identifies a unique row,even if other columns in that row have no unique values. No two rows can have the same primary key and can aslo not be assigned to null values(0).
b)Foreign key-creates links between tables which act as referential relationships. This ensures accuracy by rejecting insertions when they do not match on the referential table. They man contain duplicates and null values as well.
c)Not Null-two people may share a name or course name could be the same, but we expect no null or zero values in these entries. It ensures that important sections are filled.
d)Unique-ensures that no two or more unique identification numbers exist, for example no two people can have the sam registration number or how no two courses may have the same course ID. Which for all of these cannot be a null or zero values.
e)Check-this limits the values that can be placed in a column for example limiting the student inatake in a table. Only values that satisfy the condition are allowed in that particular column.
f)Default- sets a default value for a column incase nothing is inserted.
